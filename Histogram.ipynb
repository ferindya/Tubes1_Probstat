```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the csv file
data = pd.read_csv('anggur.csv')

# Create histogram of a numeric column
data['fixed acidity'].plot.hist(bins=10, alpha=0.5, color='blue', edgecolor='black')
plt.xlabel('value')
plt.ylabel('Frequency')
plt.title('Histogram of fixed acidity')
plt.show()
```
