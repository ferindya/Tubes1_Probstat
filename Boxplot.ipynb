```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the csv file
data = pd.read_csv('anggur.csv')

# Create boxplot of a numeric column
data.boxplot(column=['fixed acidity'])
plt.xlabel('value')
plt.title('Boxplot of fixed acidity')
plt.show()
```
